### Columns to drop:  

**adult** - no movies in this dataset are pornography    
**collection** - an id number for the collection, we already have the collection name    
**poster path** - I won't be using these images in my analysis   
**backdrop path** - T won't be using these images in my analysis   


```python

df.drop(['adult', 'collection', 'poster_path','backdrop_path'], axis=1, inplace=True)

df.info()
```

fix **datatypes** for release_date column from string to datetime

```python

df['release_date'] = pd.to_datetime(df['release_date'])
```

### Drop Rows:  
with status != released   
with null overview


```python

df.dropna(subset=['overview'], inplace=True)

df_released = df.drop(df[df['status'] != 'Released'].index).reset_index(drop=True)
df = df_released
#df['status'].value_counts()
#df['overview'].isnull().value_counts()
df.info()
```
